```r
nout = nmf(exprs(eset),5)
```

```r
w <- basis(nout)
dim(w)
loadings = lapply(1:ncol(w), function(x){loading = w[, x]
                                        loading = loading[order(loading, decreasing = TRUE)]
                                        return(loading)})
# get matrix H
h <- coef(nout)
dim(h)
```

```r
factor = 5
ggplot(data.frame(pc$x,'class' = eset$'tissue:ch1'), aes(x = PC1, y=PC2,  fill = t(h)[,factor])) + geom_point(pch =21, size = 4) + 
coord_fixed() + ggtitle(paste('factor', factor)) + scale_fill_gradientn(colors = viridis::magma(100))
```

```r
knn_idx <- FNN::get.knn(pc$x[, 1:10], k = 7)$nn.index
snn_el <- do.call(rbind, pblapply(1:nrow(knn_idx), function(x){
  nbrs <- knn_idx[x, ]
  nbr_list <- list()
  p = 1
  for(i in nbrs){
    nbr_1 <- nbrs 
    nbr_2 <- knn_idx[i, ]
    jdist <- length(intersect(nbr_1, nbr_2))/length(union(nbr_1, nbr_2))
    nbr_list[[p]] <- data.frame("node1" = x, "node2" = i, "jdist" = jdist)
    p = p+1
      }
  nbr_out <- do.call(rbind, nbr_list)
  return(nbr_out)
  
}))
snn_el = snn_el[snn_el[, 3] > 0, ]
gr <- graph_from_edgelist(as.matrix(snn_el[, 1:2]), directed = FALSE)
E(gr)$weight <- snn_el$jdist
```

```r
type = phenoData(eset)@data$source_name_ch1
library(ggraph)
set.seed(100)
lay <- layout_with_fr(gr)
pl1 <- ggraph(gr, layout = lay) + geom_edge_link(edge_colour = "grey60", edge_width = 0.5) + 
  geom_node_point(pch = 21, aes(fill = factor(type)), size = 5) +
  theme_void() + coord_fixed()
class = paste0(pData(eset)$'lupus nephritis class:ch1', "_", compartment)
class = factor(class, levels = c('normal kidney tissue_glom', 'class 3_glom', 'class 3+5_glom', 
                                      'class 4_glom', 'class 4+5_glom', 'normal kidney tissue_ti', 'class 3_ti', 
                                      'class 3+5_ti', 'class 4_ti', 'class 4+5_ti'))
pl2 <- ggraph(gr, layout = lay) + geom_edge_link(edge_colour = "grey60", edge_width = 0.5) + 
  geom_node_point(pch = 21, aes(fill = class), size = 5) +
  theme_void() + coord_fixed()

```

```r
genes_plot = c("TNF", "IL1B", "IL8")
df = do.call(rbind,lapply(genes_plot, function(x){
  df = aggregate(scale(exprs(eset)[x, ]), by = list(class), mean) 
    df$gene = x
    colnames(df) = c("class", "expression", "gene")
    return(df)
}))
df$class = factor(df$class, levels = c('normal kidney tissue_glom', 'class 3_glom', 'class 3+5_glom', 
                                      'class 4_glom', 'class 4+5_glom', 'normal kidney tissue_ti', 'class 3_ti', 
                                      'class 3+5_ti', 'class 4_ti', 'class 4+5_ti'))
df$gene = factor(df$gene, levels = genes_plot) #retain the order
pdf("figures/meija_villet_lupus_cytokines.pdf")
ggplot(df, aes(x=class, y=gene, fill = expression)) + geom_tile() + theme_bw() +  scale_fill_gradient2(
    low = 'blue', mid = 'white', high = 'red',
    midpoint = 0, guide = 'colourbar', aesthetics = 'fill', limits=c(-2,2), oob=scales::squish
  )  + coord_fixed() + theme(axis.text.x = element_text(angle = 45, hjust = 1))
dev.off()
```
