```python
# create paths and names to results folders for reference regression and cell2location models
ref_run_name = './data/cell2location/glomTI_snRNAseq_reference_signature'
```

```python
#generate a reference from the multiome data
os.chdir('/lustre/scratch126/cellgen/team297/bs16/current_projects/kidney_glomTI_response')
adata = sc.read_h5ad('./data/annotated/multiome_adata.h5ad')
```

```python
def get_counts_raw(adata):
    adata_counts = adata.uns['raw_adata'].copy()
    #adata_counts = adata_counts[adata.obs_names]
    adata = adata.raw.to_adata()
    adata.layers = adata_counts.layers.copy()
    return(adata)
adata = get_counts_raw(adata)
```

```python
adata.X = adata.layers['counts']
```

```python
#subset multiome just to gene expression
adata = adata[:, adata.var.feature_types.isin(['Gene Expression'])]
```

```python
#filter genes
from cell2location.utils.filtering import filter_genes
selected_features = filter_genes(adata, cell_count_cutoff=10, cell_percentage_cutoff2 = 0.05)
```

```python
len(selected_features) #quite a lot of features to work with
```

```python
torchmem()
```

```python
adata = adata[:, adata.var_names.isin(selected_features)]
```

```python
adata.layers['counts'] = adata.X
```

```python
# prepare anndata for the regression model
cell2location.models.RegressionModel.setup_anndata(adata=adata,
                        batch_key='donor',
                        labels_key='cell_type', layer = 'counts')
```

```python
# create the regression model
from cell2location.models import RegressionModel
mod = RegressionModel(adata)

# view anndata_setup as a sanity check
mod.view_anndata_setup()
```

```python
mod.train(max_epochs=250, use_gpu=True)
```

```python
mod.plot_history(20)
```

```python
# In this section, we export the estimated cell abundance (summary of the posterior distribution).
adata = mod.export_posterior(
    adata, sample_kwargs={'num_samples': 1000, 'batch_size': 2500, 'use_gpu': True}
)

# Save model
mod.save(f"{ref_run_name}", overwrite=True)

# Save anndata object with results
adata_file = f"{ref_run_name}/cell_to_location_reference.h5ad"
adata.write(adata_file)
```

```python
#now we work on the single cell perturbation data... 
os.chdir('/lustre/scratch126/cellgen/team297/bs16/current_projects/kidney_glomTI_response')

# create paths and names to results folders for reference regression and cell2location models
ref_run_name = './data/cell2location/glomTI_perfusion_scRNAseq_reference_signature'

#get the gene expression proportions within the target cell type
adata = sc.read_h5ad('./data/annotated/scRNAseq_perturbation_kidney.h5ad')
#get the DA cells
cells_use = pd.read_csv("data/da_results/perfusion/DE_testing_cells.csv", sep = '\t')
cells_use['stimulation'] = np.array(adata.obs.loc[cells_use.cells]['stimulation'])
cells_use['cell_type_stim'] = cells_use['cell_type'] + "_" + cells_use['stimulation']
#get cell type stim labels for our cells of interest
ct = np.array(adata.obs['cell_type'])
for i in np.unique(cells_use.cell_type_stim):
    print(i)
    cells = np.array(cells_use['cells'])[cells_use['cell_type_stim'].isin([i])] 
    ct[adata.obs_names.isin(cells)] = i
adata.obs['cell_type_stim'] = pd.Categorical(ct)
```

```python
adata =adata[~adata.obs['cell_type_stim'].isin(['GEC', "classical_monocyte", 'NK1'])] #remove non DA cells
```

```python
sc.pl.umap(adata, color = 'cell_type_stim')
```

```python
#get raw count data from adata_ref
def get_counts_raw(adata):
    adata_counts = adata.uns['raw_adata'].copy()
    adata_counts = adata_counts[adata.obs_names]
    adata = adata.raw.to_adata()
    adata.layers = adata_counts.layers.copy()
    return(adata)

adata = get_counts_raw(adata)
adata.X = adata.layers['raw_counts']
```

```python
#filter genes
from cell2location.utils.filtering import filter_genes
selected_features = filter_genes(adata, cell_count_cutoff=10, cell_percentage_cutoff2 = 0.05)
```

```python
len(selected_features) #quite a lot of features to work with
```

```python
adata = adata[:, adata.var_names.isin(selected_features)]
```

```python
adata.layers['counts'] = adata.X #this does make a new layer
```

```python
# prepare anndata for the regression model
cell2location.models.RegressionModel.setup_anndata(adata=adata,
                        batch_key='genotype', #may wish to tweak this
                        labels_key='cell_type_stim', layer = 'counts')
```

```python
# create the regression model
from cell2location.models import RegressionModel
mod = RegressionModel(adata)

# view anndata_setup as a sanity check
mod.view_anndata_setup()
```

```python
mod.train(max_epochs=250, use_gpu=True)
```

```python
mod.plot_history(20)
```

```python
# In this section, we export the estimated cell abundance (summary of the posterior distribution).
adata = mod.export_posterior(
    adata, sample_kwargs={'num_samples': 1000, 'batch_size': 2500, 'use_gpu': True}
)

# Save model
mod.save(f"{ref_run_name}", overwrite=True)

# Save anndata object with results
adata_file = f"{ref_run_name}/cell_to_location_reference.h5ad"
adata.write(adata_file)
```
