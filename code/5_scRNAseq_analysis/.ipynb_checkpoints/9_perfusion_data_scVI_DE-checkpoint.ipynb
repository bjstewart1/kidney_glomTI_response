```python
def get_counts_raw(adata):
    adata_counts = adata.uns['raw_adata'].copy()
    adata_counts = adata_counts[adata.obs_names]
    adata = adata.raw.to_adata()
    adata.layers = adata_counts.layers.copy()
    return(adata)
```

```python
os.chdir('/lustre/scratch126/cellgen/team297/bs16/current_projects/kidney_glomTI_response')
adata = sc.read_h5ad("data/annotated/scRNAseq_perturbation_kidney.h5ad")
```

```python
#subset to cells which are differentially abundant
cells_use = pd.read_csv("data/da_results/perfusion/DE_testing_cells.csv", sep = '\t')
adata = adata[adata.obs_names.isin(np.array(cells_use.cells))]
```

```python
adata = get_counts_raw(adata)
```

```python
sc.pp.filter_genes(adata, min_cells = 20) #filter genes
sc.pp.filter_genes(adata, min_counts = 100) #filter genes
```

```python
adata.obs['genotype'].value_counts()
```

```python
scvi.model.SCVI.setup_anndata(
    adata, layer="raw_counts", batch_key="genotype"
)  # prepare data for scVI
```

```python
model = scvi.model.SCVI(
    adata, gene_likelihood="nb", n_layers = 2, n_latent = 15
) 
```

```python
model.train(
    check_val_every_n_epoch=1,
    max_epochs=400,
    early_stopping=True,
    early_stopping_patience=20,
    early_stopping_monitor="elbo_validation",
)
```

```python
# Ensure convergence
import matplotlib.pyplot as plt
train_test_results = model.history["elbo_train"]
train_test_results["elbo_validation"] = model.history["elbo_validation"]
train_test_results.iloc[10:].plot(logy=True)  # exclude first 10 epochs
plt.show()
```

```python
adata.obsm["X_scVI"] = model.get_latent_representation()
sc.pp.neighbors(adata, use_rep="X_scVI")
sc.tl.umap(adata)
sc.pl.umap(adata, color=["cell_type", 'stimulation'])
```

```python
sc.tl.embedding_density(adata, basis='umap', groupby='stimulation')
```

```python
sc.pl.embedding_density(adata, basis='umap', groupby='stimulation')
```

```python
ct_list = cells_use.cell_type.unique()
```

```python
os.chdir('/lustre/scratch126/cellgen/team297/bs16/current_projects/kidney_glomTI_response/data/da_results/perfusion/response_signatures')
for ct in ct_list:
        print(ct)
        #get the stimulated cells 
        cell_idx1 = np.intersect1d(
            adata.obs_names[adata.obs["cell_type"] == ct], 
            adata.obs_names[adata.obs["stimulation"] == 'IC'])
        cell_idx1 = adata.obs_names.isin(cell_idx1)
        #get the control cells
        cell_idx2 = np.intersect1d(
            adata.obs_names[adata.obs["cell_type"] == ct], 
            adata.obs_names[adata.obs["stimulation"] == 'control'])
        cell_idx2 = adata.obs_names.isin(cell_idx2)
        de_change = model.differential_expression(idx1=cell_idx1, idx2=cell_idx2, 
                                                  batch_correction=False)
        #write them out
        fp = ct + "_differential_expression.csv"
        de_change.to_csv(fp)
```
