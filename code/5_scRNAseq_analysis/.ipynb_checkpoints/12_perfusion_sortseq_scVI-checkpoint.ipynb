```python
adata.var['highly_variable'] = True
markers = uf.get_markers(adata, 'celltype_clusters', 'counts')
```

```python
markers['4,1'].head(10)
```

```python
mrk = []
for mk in list(markers):
    mk = markers[mk].head(5)
    mk['Symbol'] = mk.index
    mrk.append(np.array(mk['Symbol']))
genes_plot = np.concatenate(mrk).ravel()
cluster_cats = list(adata.obs['celltype_clusters'].cat.categories)
genes_plot_dict = dict(zip(cluster_cats, mrk))
```

```python
sc.pl.dotplot(adata, groupby = 'celltype_clusters', var_names = genes_plot_dict, standard_scale = 'var', cmap = 'magma')
```

```python
sc.pl.matrixplot(adata, var_names = ['pct_counts_MT', 'pct_counts_in_top_50_genes', 
                                     'log1p_n_genes_by_counts', 'log1p_total_counts'], 
           groupby = 'celltype_clusters', standard_scale = 'var', swap_axes = True, dendrogram = False)
```

```python
celltype_map = {
   '0' : 'NK1',
    '1' : 'CD8_T_cell',
    '2' : 'classical_monocyte', 
    '3' : 'classical_monocyte',
    '4,0' : 'classical_monocyte',
    '4,1' : 'macrophage',
    '5' : 'nonclassical_monocyte',
    '6' : 'GEC',
    '7' : 'CD4_T_cell',
    '8' : 'NK1',
    '9,0' : 'distal_tubule',
    '9,1' : 'PTEC',
    '10' : 'low_quality'
} 
# add a new `.obs` column called `cell type` by mapping clusters to annotation using pandas `map` function
adata.obs['cell_type'] = adata.obs['celltype_clusters'].map(celltype_map).astype('category')
```

```python
adata = adata[~adata.obs['cell_type'].isin(['low_quality'])]
```

```python
sc.pp.neighbors(adata, use_rep= 'X_scVI', n_neighbors = 10)
```

```python
sc.tl.umap(adata)
```

```python
sc.set_figure_params(frameon=False, dpi=150, figsize=(5,5))

sc.pl.umap(adata, color = 'cell_type', legend_loc = 'on data', add_outline = True)
```

```python
#make the cell types categorical
ct_categories = [
'classical_monocyte', 'nonclassical_monocyte', 'macrophage', 'CD4_T_cell', 'CD8_T_cell', 'NK1', 'PTEC', 'distal_tubule', 'GEC'
]
adata.obs['cell_type'] = pd.Categorical(adata.obs['cell_type'], categories=ct_categories)
```
