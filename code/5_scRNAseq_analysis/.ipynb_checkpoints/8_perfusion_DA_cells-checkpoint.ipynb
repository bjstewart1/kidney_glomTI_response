```r
#in this notebook we get the differentially abundant cells for performing downstream analysis
#in this notebook we generate a metric to determine how perturbed each cell type is.... 
project_dir = "~/ben_lustre/current_projects/kidney_glomTI_response"
setwd(project_dir)
suppressPackageStartupMessages(source("https://raw.githubusercontent.com/bjstewart1/helpful_singlecell/main/functions.R"))
ad <- import("anndata")
sc <- import("scanpy")
scv <- import("scvelo")
sp = import("scipy")
clean_theme <- theme_bw() + theme(axis.text = element_text(color = "black"), axis.title = element_text(color = "black"))
```

```r
setwd("~/lustre/scratch126/cellgen/team297/bs16/current_projects/kidney_glomTI_response")
adata = sc$read_h5ad('./data/annotated/scRNAseq_perturbation_kidney.h5ad')
```

```r
da_results = readRDS("data/da_results/perfusion/da_results.RDS")
milo = readRDS("data/da_results/perfusion/perfusion_milo.RDS")
```

```r
#get cells for DE - here we grab cells from differentially abundant neighborhoods for doing DE testing and signalling analysis
#for each cell type we get significantly up and down regulated neighborhoods and save these (omiting duplicates for downstream analysis)
cells_test = c('classical_monocyte', 'NK1', 'GEC')
cells_test = do.call(rbind, lapply(cells_test, function(ct){
message(ct)
down_neighborhoods = rownames(da_results[da_results$logFC < 0 & da_results$SpatialFDR < 0.1 , ])
down_cells = Matrix::rowSums(milo@nhoods[, as.numeric(down_neighborhoods)]) > 0
ct_cells = colData(milo)$cell_type %in% ct
stim_cells = colData(milo)$stimulation %in% 'IC'
down_cells = rownames(colData(milo)[down_cells & ct_cells & stim_cells
                                    , ])
#do up - these are control cells
up_neighborhoods = rownames(da_results[da_results$logFC > 0 & da_results$SpatialFDR < 0.1, ])
up_cells = Matrix::rowSums(milo@nhoods[, as.numeric(up_neighborhoods)]) > 0
ct_cells = colData(milo)$cell_type %in% ct
contr_cells = colData(milo)$stimulation %in% 'control'
up_cells = rownames(colData(milo)[up_cells & ct_cells & contr_cells
                                  , ])
#get all cells to test
cells_use = c(down_cells, up_cells)
return(data.frame("cell_type" = ct, "cells" = cells_use) )  
}))
```

```r
cells_test = cells_test[!duplicated(cells_test$cells), ]
write.table(cells_test, "data/da_results/perfusion/DE_testing_cells.csv", sep = '\t')
```

```r
obs  = adata$obs
rownames(obs) = adata$obs_names$values
stimtabs = lapply(unique(cells_test$cell_type), function(ct){
   cells = cells_test[cells_test$cell_type %in% ct, 'cells']
return(    table(obs[cells, 'stimulation'])
)    
})
names(stimtabs) = unique(cells_test$cell_type)
```

```r
#get a plot of DA neighborhoods of interest
cells_test = c('classical_monocyte', 'NK1', 'GEC')
da_results$NhoodGroup = NA
for(ct in cells_test){
da_results[da_results$cell_type %in% ct & da_results$SpatialFDR < 0.1 & da_results$logFC > 0, 'NhoodGroup'] = paste0(ct, "_", 'control')
da_results[da_results$cell_type %in% ct & da_results$SpatialFDR < 0.1 & da_results$logFC < 0, 'NhoodGroup'] = paste0(ct, "_", 'IC')
}
```

```r
plotNhoodGraphDA(milo, da_results, layout="X_umap",alpha=0.1) + coord_fixed()
```

```r
nhood_group_colors = c("GEC_control" = 'forestgreen', "GEC_IC" = "pink", 
                      "NK1_control" = 'purple', "NK1_IC" = 'orangered',
                      "classical_monocyte_control" = 'royalblue', "classical_monocyte_IC" = 'chocolate1')
```

```r
plotNhoodGroups(milo, da_results, layout="X_umap",alpha=0.1) + coord_fixed() + scale_fill_manual(values = nhood_group_colors, na.value = 'grey90')
```
