```r
library(nichenetr) #niche net is used for inferring ligand activities in cells, which can help nominate important ligand receptor interactions 
library(dplyr)
```

```r
#use niche net to nominate ligands and receptors active in our cell types particularly GECS..... 
adata = sc$read_h5ad("./data/annotated/scRNAseq_perturbation_kidney.h5ad")
#make raw -  full matrix to use..
adata = adata$raw$to_adata()
#subset to response cells 
response_cells = read.csv('./data/da_results/perfusion/DE_testing_cells.csv', sep = '\t')
adata = adata[adata$obs_names$values %in% response_cells$cells]
#establish stimulated vs unstimulated cells
adata$obs$cell_type_stim = paste0(adata$obs$cell_type, "_", adata$obs$stimulation)
```

```r
sce = SingleCellExperiment(assays = list("logcounts" = as.matrix(t(adata$X))), 
                          colData = adata$obs, rowData = adata$var)
```

```r
#download ligand receptor network and the ligand-target matrix
nichenet_dir = './data/public_data/nichenet'
lr_network = readRDS(file.path(nichenet_dir, "lr_network_human_21122021.rds"))
ligand_target_matrix = readRDS(file.path(nichenet_dir, "ligand_target_matrix_nsga2r_final.rds"))
```

```r
get_expressed_genes = function(celltypes, celltype_label, pct = 0.2, adata){
    ct_label = adata$obs[, celltype_label]
    
    genes = unique(unlist(lapply(celltypes, function(ct){
        idx = ct_label %in% ct
        ncells_expressing =  Matrix::colSums(adata$X[idx, ] > 0)
        total_cells = sum(idx)
        expressed_genes = adata$var_names$values[ncells_expressing/total_cells > pct]
        return(expressed_genes)
    })))
   return(genes) 
}
```

```r
response_genesets = read.csv('./data/da_results/perfusion/response_signatures/DEG_use.csv', row.names = 1)
```

```r
#run nichenet...

#get expressed genes in receiver cells
receiver = 'GEC_IC'
sender = c("classical_monocyte_IC", "NK1_IC")

#get genes expressed in receiver cells (GECS)
receiver_expressed_genes = get_expressed_genes(celltypes = receiver, celltype_label = 'cell_type_stim', pct = 0.2, adata = adata)

#get genes expressed in sender cells (mono & NK1)
sender_expressed_genes = get_expressed_genes(celltypes = sender, celltype_label = 'cell_type_stim', pct = 0.2, adata = adata)

#for expressed ligands we are looking for ligands which are up in the OTHER stimulated cells (i.e. immune cells signalling to stimulated GECS)
#get our upregulated genesets from DE analysis
genesets = read.csv('./data/da_results/perfusion/response_signatures/DEG_use.csv', row.names = 1)
#select stimulated GECs
stimulated_geneset = response_genesets[response_genesets$cell_type %in% 'GEC', 'X'] #this is response geneset in GECS
stimulated_immune_cell_geneset = unique(response_genesets[response_genesets$cell_type %in% c("NK1", "classical_monocyte"), 'X'])

#get ligands and receptors from the nichenet resource
ligands <- lr_network %>% pull(from) %>% unique()
receptors <- lr_network %>% pull(to) %>% unique()
#expressed_ligands <- intersect(ligands, sender_expressed_genes)  #look at all genes expressed by senders 
expressed_ligands <- intersect(ligands, stimulated_immune_cell_geneset)  #look at upregulated ligands
expressed_receptors <- intersect(receptors, receiver_expressed_genes) #look at all genes expressed by receivers    

#get potential ligands
potential_ligands <- lr_network %>% 
  filter(from %in% expressed_ligands & to %in% expressed_receptors) %>%
  pull(from) %>% unique()    

#for this we use the genes that are expressed at all in stimulated GECs
background_genes = receiver_expressed_genes

#get ligand activities
ligand_activities <- predict_ligand_activities(geneset = stimulated_geneset, 
                                               background_expressed_genes = background_genes,
                                               ligand_target_matrix = ligand_target_matrix,
                                               potential_ligands = potential_ligands)

#rank the ligand activities
ligand_activities <- ligand_activities %>% 
  arrange(-aupr) %>% 
  mutate(rank = rank(desc(aupr)))
#get the top ligands
top_ligands <- ligand_activities %>%
  top_n(20, aupr) %>%  #take the top 20 ligands
  arrange(-aupr) %>%
  pull(test_ligand) %>%
  unique()

#get ligand target potential
ligand_target_potential = do.call(rbind, lapply(top_ligands, function(x){
    get_weighted_ligand_target_links(x, geneset = stimulated_geneset, ligand_target_matrix = ligand_target_matrix, n = 500)
}))
#this gives a matrix of ligand target links
active_ligand_target_links <- 
  prepare_ligand_target_visualization(ligand_target_df = ligand_target_potential, 
                                      ligand_target_matrix = ligand_target_matrix)
#for visualisation we mung this into an ordered form
order_ligands = rev(intersect(top_ligands, colnames(active_ligand_target_links)))
order_targets = intersect(unique(ligand_target_potential$target), rownames(active_ligand_target_links))
vis_ligand_target = t(active_ligand_target_links[order_targets, order_ligands])
#subset so that we keep only columns where at least one gene has a regulatory potential >= 0.05
vis_ligand_target = vis_ligand_target[, colSums(vis_ligand_target >= 0.05) > 1]
#get a nice output list
nichenet_out = list(
'ligand_target_links' = active_ligand_target_links,
'vis_ligand_target' = vis_ligand_target,
'ligand_target_potential' = ligand_target_potential,
'ligand_activities' = ligand_activities,
'top_ligands' = top_ligands    
)
lapply(names(nichenet_out), function(x){
    write.csv(nichenet_out[[x]], file.path('data/da_results/perfusion/nichenet', paste0(x, '.csv')), sep = '\t')
})
```
