```python
#perfusion geneset signalling analysis 
import os
os.chdir('/lustre/scratch126/cellgen/team297/bs16/current_projects/kidney_glomTI_response/code')
import anndata as ad
import scanpy as sc
import os
import pandas as pd
import numpy as np
import useful_functions as uf
import scvi
sc.set_figure_params(figsize=(6, 6), dpi = 150, fontsize = 10)
import sys
print(sys.executable)
import matplotlib as mpl
import matplotlib.pyplot as plt

from datetime import datetime
print(datetime.now(tz=None))

#now we work on the single cell perturbation data... 
os.chdir('/lustre/scratch126/cellgen/team297/bs16/current_projects/kidney_glomTI_response')

# create paths and names to results folders for reference regression and cell2location models
ref_run_name = './data/cell2location/glomTI_perfusion_scRNAseq_reference_signature'

#get the gene expression proportions within the target cell type
adata = sc.read_h5ad('./data/annotated/scRNAseq_perturbation_kidney.h5ad')
#get the DA cells
cells_use = pd.read_csv("data/da_results/perfusion/DE_testing_cells.csv", sep = '\t')
cells_use['stimulation'] = np.array(adata.obs.loc[cells_use.cells]['stimulation'])
cells_use['cell_type_stim'] = cells_use['cell_type'] + "_" + cells_use['stimulation']
#get cell type stim labels for our cells of interest
ct = np.array(adata.obs['cell_type'])
for i in np.unique(cells_use.cell_type_stim):
    print(i)
    cells = np.array(cells_use['cells'])[cells_use['cell_type_stim'].isin([i])] 
    ct[adata.obs_names.isin(cells)] = i
adata.obs['cell_type_stim'] = pd.Categorical(ct)
```

```python
adata =adata[~adata.obs['cell_type_stim'].isin(['GEC', "classical_monocyte", 'NK1'])] #remove non DA cells
```

```python
sc.pl.umap(adata, color = 'cell_type_stim')
```

```python
#make a dict of relevant genes to make our nice plots 
gec_response_dict = {'ligands' : ['IL6', 'CXCL1', 'CXCL2', 'CXCL3', 'CXCL8', 'CSF3', 'CCL2', 'TGFB3', 'TGFB2', 'KITLG', 'HBEGF'],
'receptors' : ['IL2RA', 'IL15RA', 'TNFRSF1A', 'TACR1', 'PLAUR', 'SDC4'],
'adhesion' : ['SELE', 'ICAM1', 'ICAM4', 'JAM3','PVR', 'ITGB3', 'NINJ1', 'CD83', 'CD200', 'ITGA9', 'CD55'],
'signal_transduction' : ['IRAK2', 'TNIP1', 'TNFAIP1', 'TNFAIP2', 'TNFAIP3','TRAF1', 'TRAF3', 'MAP2K3', 'IRF1', 'SMAD3', 'NFIL3', 'NRIP1'],
'heatshock' : ['HSPA1A', 'HSPA1B', 'HSPA6', 'CRYAB', 'HSPA8','DNAJA1', 'DNAJA4', 'HSPE1', 'HSPD1', 'DNAJB4', 'HSP90AA1', 'HSPB1', 'HSP90AB1', 'DNAJB6', 'HSPA9'],
'TFs' : ['NFKB1', 'NFKB2', 'NFKBIA', 'NR4A2', 'NFAT5', 'BCL6', 'BCL3', 'BHLHE40', 'ATF3', 'SOX7', 'IER5', 'FOSL1', 'CREB5', 'ETS1'],
}
```

```python
sc.pl.dotplot(adata[adata.obs.cell_type_stim.isin(['GEC_IC', 'GEC_control', 'classical_monocyte_IC', 'classical_monocyte_control',
                                                   'NK1_control', 'NK1_IC'])], groupby = 'cell_type_stim', var_names = gec_response_dict, 
              swap_axes = False, dot_min = 0.1,
              standard_scale = 'var', save = 'GEC_response_gene_heatmap.pdf', cmap = 'magma')
```

```python
#get the stuff that is thought to be downstream of IL1B by the nichenet prediction
ligand_potential = pd.read_csv("./data/da_results/perfusion/nichenet/ligand_target_potential.csv", index_col = 0)
ligand_potential = ligand_potential[ligand_potential.ligand.isin([ 'IL1B', 'IL1A', 'TNF'])]
ligand_potential = ligand_potential[ligand_potential.weight >0]
tnf_il1b_influenced = np.unique(ligand_potential.target)
```

```python
subset_dict = {}
# Iterate through the dictionary items and their values
for category, values in gec_response_dict.items():
    common_values = [value for value in values if value in tnf_il1b_influenced]
    if common_values:
        subset_dict[category] = common_values
subset_dict.pop('heatshock')
```

```python
sc.pl.dotplot(adata[adata.obs.cell_type_stim.isin(['GEC_IC', 'GEC_control', 'classical_monocyte_IC', 'classical_monocyte_control',
                                                   'NK1_control', 'NK1_IC'])], groupby = 'cell_type_stim', var_names = subset_dict, 
              swap_axes = False, dot_min = 0.1,
              standard_scale = 'var', cmap = 'magma', save = 'GEC_response_gene_heatmap_IL1B_TNF.pdf')
```
